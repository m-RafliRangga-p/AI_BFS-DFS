# **BFS Algorithm**Blank diagram.jpeg

**Inisialisasi BFS Search**
- bfs_search(graph, start, target): Fungsi yang menerima tiga parameter:
    - graph: Representasi graf dari struktur direktori.
    - start: Node awal untuk memulai pencarian (root dari direktori).
    - target: File yang dicari.
- visited: Set untuk melacak node yang telah dikunjungi.
- queue: Antrian untuk melacak node yang akan dikunjungi, diinisialisasi dengan node awal start.

**Algoritma BFS**
- while queue:: Loop berjalan selama antrian tidak kosong.
- node = queue.popleft(): Mengambil node dari antrian.
- if node not in visited:: Memeriksa apakah node sudah dikunjungi. Jika belum:
    - print(f"Visiting: {node}"): Mencetak node yang sedang dikunjungi.
    - visited.add(node): Menambahkan node ke set visited.
    - if node == target:: Memeriksa apakah node yang sedang dikunjungi adalah target. Jika ya:
        - print(f"Found {target}!"): Mencetak pesan bahwa target ditemukan.
        - return True: Mengakhiri fungsi dan mengembalikan True.

**Neighbor**
- if node in graph:: Memeriksa apakah node ada dalam graf sebelum mencoba mengakses tetangganya. Hal ini mencegah KeyError.
- for neighbor in graph[node]:: Mengiterasi setiap tetangga dari node saat ini.
    - if neighbor not in visited:: Memeriksa apakah tetangga belum dikunjungi.
    - queue.append(neighbor): Menambahkan tetangga ke antrian untuk dikunjungi.

return False: Jika antrian kosong dan target tidak ditemukan, fungsi mengembalikan False.

```python
from collections import deque

def bfs_search(graph, start, target):
    visited = set()
    queue = deque([start])

    while queue:
        node = queue.popleft()
        if node not in visited:
            print(f"Visiting: {node}")
            visited.add(node)
            if node == target:
                print(f"Found {target}!")
                return True
            # Periksa apakah node ada dalam graf sebelum mencoba mengakses tetangganya
            if node in graph:
                for neighbor in graph[node]:
                    if neighbor not in visited:
                        queue.append(neighbor)
    return False

# Contoh penggunaan dengan nama file
graph = {
    '/': ['/lib', '/bin', '/user'],
    '/lib': [],
    '/bin': [],
    '/user': ['/home', '/program'],
    '/home': ['/documents', '/download', 'file.txt', 'foto.png'],
    '/program': ['word.exe', 'excel.exe', 'powerpoint.exe'],
    '/documents': ['/file', 'doc1.docx', 'excel1.xlsx', 'ppt1.pdf'],
    '/download': ['/film', 'manga.pdf', 'data.xlsx', 'kiky.pdf'],
    '/file': [],
    '/film': ['Godzila.mp4']
}

# Mencari Godzila.mp4
file_to_find = 'Godzila.mp4'
found = bfs_search(graph, '/', file_to_find)

if not found:
    print(f"{file_to_find} not found in the graph.")
```

# **DFS Algorithm**

1. **Deklarasi DFS Search**
    - dfs_search(graph, node, target, visited=None): Fungsi yang menerima empat parameter:
        - graph: Representasi graf dari struktur direktori.
        - node: Node saat ini yang sedang dievaluasi.
        - target: File yang dicari.
        - visited: Set yang melacak node yang telah dikunjungi.
    - visited: Set untuk melacak node yang telah dikunjungi.
    - DFS Rekursif: Fungsi ini menggunakan pendekatan rekursif untuk menjelajahi graf.
    - Base Case: Jika node belum dikunjungi dan bukan merupakan file yang dicari, fungsi akan melanjutkan pencarian ke tetangga-tetangganya.
2. **Algoritma DFS**
    - Fungsi dfs_search melakukan penelusuran secara rekursif dari node ke tetangga-tetangganya sampai file yang dicari ditemukan atau seluruh graf telah diperiksa.
3. **Iterasi Melalui Tetangga**
    - Algoritma DFS secara rekursif menelusuri setiap tetangga dari node saat ini.
    - Tiap kali menelusuri tetangga, node yang sedang dikunjungi dicetak untuk memantau proses pencarian.
4. **Penemuan Target**
    - Jika target (file yang dicari) ditemukan, pesan "Found Godzila.mp4!" dicetak.
5. **Akhir Pencarian**
    - Jika target tidak ditemukan setelah menelusuri seluruh graf, pesan "Godzila.mp4 not found in the graph." dicetak.

```python
def dfs_search(graph, node, target, visited=None):
    if visited is None:
        visited = set()
    if node not in visited:
        print(f"Visiting: {node}")
        visited.add(node)
        if node == target:
            print(f"Found {target}!")
            return True
        if node in graph:  # Pastikan node ada dalam graf sebelum mengakses tetangganya
            for neighbor in graph[node]:
                if dfs_search(graph, neighbor, target, visited):
                    return True
    return False

# Contoh penggunaan dengan nama file
graph = {
    '/': ['/lib', '/bin', '/user'],
    '/lib': [],
    '/bin': [],
    '/user': ['/home', '/program'],
    '/home': ['/documents', '/download', 'file.txt', 'foto.png'],
    '/program': ['word.exe', 'excel.exe', 'powerpoint.exe'],
    '/documents': ['/file', 'doc1.docx', 'excel1.xlsx', 'ppt1.pdf'],
    '/download': ['/film', 'manga.pdf', 'data.xlsx', 'kiky.pdf'],
    '/file': [],
    '/film': ['Godzila.mp4']
}

# Mencari file
file_to_find = 'Godzila.mp4'
found = dfs_search(graph, '/', file_to_find)

if not found:
    print(f"{file_to_find} not found in the graph.")
```
